```python
import tiktoken
import json
import pandas as pd
import pickle
from tiktoken._educational import SimpleBytePairEncoding #run in Google Colab, it might not work in Jupyter
```

```python
def read_text_from_json(df):
    texts = []
    line_count = 0  # Initializing line counter
    for index, row in df.iterrows():
        text = row['text']
        if not isinstance(text, str):  # Checking if the element is not a string
            print(f"Non-string element: {text}, Type: {type(text)}")
        else:
            texts.append(text)  # Adding to the list only if it's a string
            line_count += 1
            if line_count >= 10:
                break
    return texts

# Mounting Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Specifying the full path to the file
file_path = '/content/drive/MyDrive/wiki_arm.json'

# Reading the JSON file into a pandas DataFrame
df = pd.read_json(file_path, orient='columns', compression='infer')

# Calling the modified function to read the first 10 lines
texts = read_text_from_json(df)
```

```python
print(texts)
```

```python
# Joining all the texts into a single string
joined_text = "\n".join(texts)
print(joined_text)
```

```python
# Training the BPE tokenizer with vocabulary size 400
enc = SimpleBytePairEncoding.train(joined_text, vocab_size=400, pat_str='.')

# Saving the trained BPE tokenizer to a file
with open('bpe_tokenizer_new.pkl', 'wb') as f:
    pickle.dump(enc, f)
```

```python
# Loading the trained BPE tokenizer from the file
with open('bpe_tokenizer_new.pkl', 'rb') as f:
    enc_loaded = pickle.load(f)

# Example text to encode
text_to_encode = "ิตีฝ ิฑีถีถีกีถ ีฅีด, ีฝีธีพีธึีธึีด ีฅีด ีีกีตีกีฝีฟีกีถีซ ิฑีดีฅึีซีฏีตีกีถ ีฐีกีดีกีฌีฝีกึีกีถีธึีดี ีีพีตีกีฌีกีฃีซีฟีธึีฉีตีกีถ ีขีกีชีถีธึีด: "

# Encoding the text using the loaded tokenizer
encoded_text = enc_loaded.encode(text_to_encode)

# Decoding the encoded text using the loaded tokenizer
decoded_text = enc_loaded.decode(encoded_text)

# Print the results
print("Encoded text:", encoded_text)
print("Decoded text:", decoded_text)
```

```python
# Accessing the resulting vocabulary (mergeable ranks)
mergeable_ranks = enc_loaded.mergeable_ranks

# Print the resulting vocabulary
print("Resulting vocabulary (mergeable ranks):")
for token_bytes, rank in mergeable_ranks.items():
    try:
        token_str = token_bytes.decode('utf-8')
    except UnicodeDecodeError:
        token_str = None  # Skip if unable to decode as UTF-8
    if token_str is not None:
        print(f"Token: {token_str}  Rank: {rank}")
    else:
        print(f"Token: {token_bytes.hex()}  Rank: {rank} (Not UTF-8)")
```
